## Merge all DFs with its respective IDs

```python
transaction_merged_brand_df = pd.merge(df, brand[['product_brand','brand_id']],on='product_brand')

transaction_brand_merged_category_df = pd.merge(transaction_merged_brand_df,category[['product_category','category_id']], on='product_category')

transaction_brand_category_merged_type_df = pd.merge(transaction_brand_merged_category_df, product_type[['product_type','product_type_id']], on='product_type')

transaction_brand_category_type_merged_name_df = pd.merge(transaction_brand_category_merged_type_df, product_name[['product_name','product_name_id']], on='product_name')

all_merged_df = transaction_brand_category_type_merged_name_df.copy()

all_merged_df.columns
```

# Transactions

```python
# Create separate dataframe for Transactions table columns
transactions = pd.DataFrame({
    'transaction_id': all_merged_df['transaction_id'],
    'customer_id': all_merged_df['customer_id'],
    'brand_id':all_merged_df['brand_id'],
    'category_id':all_merged_df['category_id'],
    'product_type_id':all_merged_df['product_type_id'],
    'product_name_id':all_merged_df['product_name_id'],
    'date': all_merged_df['date'],
    'year': all_merged_df['year'],
    'month': all_merged_df['month'],
    'time': all_merged_df['time'],
    'total_purchases': all_merged_df['total_purchases'],
    'amount': all_merged_df['amount'],
    'total_amount': all_merged_df['total_amount'],
    'payment_method':all_merged_df['payment_method'],
    'order_status': all_merged_df['order_status'],
    'shipping_method':all_merged_df['shipping_method'],
    'feedback': all_merged_df['feedback'],
    'ratings': all_merged_df['ratings']
})
```
