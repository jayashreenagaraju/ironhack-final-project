## Fix duplicate customer ids

```python
# Step 1: Identify duplicates in customer_id
duplicates = df[df.duplicated('customer_id', keep=False)]

# Step 2: Generate unique customer_id for duplicates
new_ids = {}
for idx, row in duplicates.iterrows():
    if row['customer_id'] not in new_ids:
        new_ids[row['customer_id']] = [row['customer_id']]
    new_ids[row['customer_id']].append(f"{row['customer_id']}{len(new_ids[row['customer_id']])}")

# Apply new unique customer_ids
for idx, row in duplicates.iterrows():
    df.loc[idx, 'customer_id'] = new_ids[row['customer_id']].pop(0)

# Verify the result
df.shape
```

## Fix duplicate transaction IDs

```python
# Step 1: Identify duplicates in transaction_id
duplicates = df[df.duplicated('transaction_id', keep=False)]

# Step 2: Generate unique transaction_id for duplicates
new_ids = {}
for idx, row in duplicates.iterrows():
    if row['transaction_id'] not in new_ids:
        new_ids[row['transaction_id']] = [row['transaction_id']]
    new_ids[row['transaction_id']].append(f"{row['transaction_id']}{len(new_ids[row['transaction_id']])}")

# Apply new unique transaction_id
for idx, row in duplicates.iterrows():
    df.loc[idx, 'transaction_id'] = new_ids[row['transaction_id']].pop(0)

# Verify the result
df.shape
```

```python
def fill_date(row):
    if pd.isna(row['date']):
        year = str(row['year'])
        month = row['month']
        day = '20'  # You can change '20' as per your requirement
        month_str = str(pd.to_datetime(month, format='%B').month).zfill(2)
        return f"{year}-{month_str}-{day}"
    else:
        return row['date']

# Apply the function to fill NaN values in 'date' column
df['date'] = df.apply(fill_date, axis=1)

# Ensure we have complete months for both 2022 and 2023

# Extract unique years and months from the DataFrame
years = df['year'].unique()
months = df['month'].unique()

# Create a new DataFrame with all combinations of year and month
complete_data = []

for year in years:
    for month in months:
        complete_data.append({
            'year': year,
            'month': month
        })

complete_df = pd.DataFrame(complete_data)

# Merge to fill in missing months for 2023
df = pd.merge(complete_df, df, on=['year', 'month'], how='left')

# Apply the function again to ensure all NaN values are filled
df['date'] = df.apply(fill_date, axis=1)

# Now df should have complete month information for both 2022 and 2023
display(df)
```

```python
# Function to fill NaN values in 'date' column
def fill_date(row):
    if pd.isna(row['date']):
        year = str(row['year'])
        month = row['month']
        day = '20'
        # Format month as two digits (e.g., '01' for January)
        month_str = str(pd.to_datetime(month, format='%B').month).zfill(2)
        return f"{year}-{month_str}-{day}"
    else:
        return row['date']

# Apply the function to fill NaN values in 'date' column
df['date'] = df.apply(fill_date, axis=1)
```

```python
from dateutil.relativedelta import relativedelta
# Define a mapping from month names to month numbers
month_mapping = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4,
    'May': 5, 'June': 6, 'July': 7, 'August': 8,
    'September': 9, 'October': 10, 'November': 11, 'December': 12
}

# Convert the date column to datetime
df['date'] = pd.to_datetime(df['date'])

# Function to update the date with the correct month
def correct_month(row):
    correct_month_number = month_mapping[row['month']]
    # Get the current date
    current_date = row['date']
    # Use relativedelta to safely change the month
    new_date = current_date + relativedelta(month=correct_month_number)
    return new_date

# Apply the function to each row
df['date'] = df.apply(correct_month, axis=1)

display(df)
```
