# K-means clustering

## Handle Categorical Variables

```python
# Identify categorical and numerical columns
categorical_cols = ['name', 'email', 'address', 'city', 'state', 'country', 'gender', 'income', 'customer_segment', 'month', 'time', 'product_category', 'product_brand', 'product_type', 'feedback', 'shipping_method', 'payment_method', 'order_status', 'product_name']
numerical_cols = ['age', 'total_purchases', 'amount', 'total_amount', 'ratings']

# Preprocessing for numerical data
numerical_transformer = StandardScaler(with_mean=False)

# Preprocessing for categorical data
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

# Combine preprocessors
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])
```

## Normalise data

```python
# Create preprocessing and kmeans pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('scaler', StandardScaler(with_mean=False))])
```

## Select Features and apply K-means

```python
features = df.drop(columns=['transaction_id', 'customer_id', 'date', 'year', 'zipcode', 'phone'])
preprocessed_data = pipeline.fit_transform(features)

# Determine the optimal number of clusters using the elbow method
sse = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(preprocessed_data)
    sse.append(kmeans.inertia_)

plt.plot(k_range, sse)
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.show()
```

## Fit K-means with Optimal Number of Clusters

```python
# Choose an appropriate number of clusters, for example, k=4
kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(preprocessed_data)

# Add cluster labels to the original dataframe
df['cluster'] = kmeans.labels_
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import TruncatedSVD
from joblib import Parallel, delayed
import matplotlib.pyplot as plt

# Assuming preprocessed_data is a csr_matrix
n_components = 100  # Number of components for SVD
svd = TruncatedSVD(n_components=n_components)
reduced_data = svd.fit_transform(preprocessed_data)

# Sample the reduced data
sampled_indices = np.random.choice(reduced_data.shape[0], size=int(0.1 * reduced_data.shape[0]), replace=False)
sampled_data = reduced_data[sampled_indices, :]

# Function for parallel computation
def compute_silhouette(k):
    kmeans = KMeans(n_clusters=k, random_state=0)
    kmeans.fit(sampled_data)
    return silhouette_score(sampled_data, kmeans.labels_)

# Compute silhouette scores in parallel
k_range = range(2, 11)
silhouette_scores = Parallel(n_jobs=-1)(delayed(compute_silhouette)(k) for k in k_range)

# Print silhouette scores
print(silhouette_scores)

# Plot silhouette scores
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Different Numbers of Clusters')
plt.show()
```

```python
# Fit K-means with the optimal number of clusters (k = 8)
optimal_k = 8
kmeans = KMeans(n_clusters=optimal_k, random_state=0)
kmeans.fit(reduced_data)
# Assign the cluster labels to the original data
cluster_labels = kmeans.labels_
# Print the cluster centroids
centroids = kmeans.cluster_centers_
print("Cluster centroids:\n", centroids)
```

```python
# Reduce the data to 2 dimensions using PCA for visualization
pca = TruncatedSVD(n_components=2)
data_2d = pca.fit_transform(reduced_data)

# Plot the clusters
plt.figure(figsize=(10, 6))
plt.scatter(data_2d[:, 0], data_2d[:, 1], c=cluster_labels, cmap='viridis', marker='o')
plt.title('K-means Clustering with 8 Clusters')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Compute the WCSS
wcss = kmeans.inertia_
print("Within-Cluster Sum of Squares (WCSS):", wcss)
```

## Analyze and Interpret the Clusters

### Examine Cluster Centers

```python
# Convert cluster centers back to original scale
inverse_transformed_centers = preprocessor.named_transformers_['num'].inverse_transform(kmeans.cluster_centers_[:, :len(numerical_cols)])
centers_df = pd.DataFrame(inverse_transformed_centers, columns=numerical_cols)
display(centers_df)
```

### Profile Clusters

```python
# Select only numeric columns
numeric_columns = df.select_dtypes(include=[int, float]).columns
cluster_profiles = df.groupby('cluster')[numeric_columns].mean()
display(cluster_profiles)
```

```python
# Assign descriptive labels to clusters
df['cluster_label'] = df['cluster'].map({
    0: 'High Value Customers',
    1: 'Low Value Customers',
    2: 'Frequent Shoppers',
    3: 'Occasional Shoppers',
    4: 'Big Spenders',
    5: 'Discount Seekers',
    6: 'Impulse Buyers',
    7: 'Seasonal shoppers'
})
display(df[['customer_id', 'cluster', 'cluster_label']].head())
```

## Visualize the Clusters

### 2D Visualization with PCA

```python
from sklearn.decomposition import PCA
import seaborn as sns

pca = PCA(n_components=2)
pca_result = pca.fit_transform(preprocessed_data)
df['pca_one'] = pca_result[:,0]
df['pca_two'] = pca_result[:,1]

plt.figure(figsize=(16,10))
sns.scatterplot(
    x="pca_one", y="pca_two",
    hue="cluster",
    palette=sns.color_palette("hsv", len(df['cluster'].unique())),
    data=df,
    legend="full",
    alpha=0.3
)
plt.title('2D PCA of Clusters')
plt.show()
```
